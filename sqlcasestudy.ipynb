```python
response = requests.get("https://api.spacexdata.com/v4/launches")
data = response.json()
df = pd.DataFrame(data)
launch_df = df[['name', 'date_utc', 'success', 'details', 'rocket']]
launch_df
```

```python
df['date_utc'] = pd.to_datetime(df['date_utc'])
df['year'] = df['date_utc'].dt.year
```

```python
df
```

```python
response1 = requests.get("https://api.spacexdata.com/v4/rockets")
data1 = response1.json()
df1= pd.DataFrame(data1)
df1
```

```python
df1[['id', 'name', 'type', 'active', 'stages']].head()
```

```python
df1 = df1.rename(columns={'id': 'rocket', 'name': 'rocket_name'})
merged_df = pd.merge(df, df1, on='rocket', how='left')
merged_df
```

```python
import numpy as np
countries = ['USA', 'Russia', 'India', 'China', 'France']
merged_df['country'] = np.random.choice(countries, size=len(merged_df))
merged_df[['name', 'rocket_name', 'country']]
```

```python
import json


for col in merged_df.columns:
    if merged_df[col].apply(lambda x: isinstance(x, (dict, list))).any():
        merged_df[col] = merged_df[col].apply(lambda x: json.dumps(x) if isinstance(x, (dict, list)) else x)


import sqlite3

conn = sqlite3.connect("spacex_launches.db")
merged_df.to_sql('launches', conn, if_exists='replace', index=False)


import pandas as pd
pd.read_sql("SELECT * FROM launches LIMIT 5;", conn)

```

```python
query1 = """
SELECT country, COUNT(*) AS launch_count
FROM launches
GROUP BY country
ORDER BY launch_count DESC;
"""
pd.read_sql(query1, conn)
```

```python
query2 = """
SELECT year, COUNT(*) AS launch_count
FROM launches
GROUP BY year
ORDER BY launch_count DESC
LIMIT 1;
"""
pd.read_sql(query2, conn)
```

```python
query3 = """
SELECT name AS mission_name, COUNT(*) AS times_launched
FROM launches
GROUP BY mission_name
ORDER BY times_launched DESC
LIMIT 5;
"""
pd.read_sql(query3, conn)
```
